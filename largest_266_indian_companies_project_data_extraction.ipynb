```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page=requests.get('https://disfold.com/india/companies/')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
table1=soup.find_all('table',class_='striped')
table1
```

```python
all_data=[]
for i in table1 :
  for j in i.find_all("td") :
    all_data.append(j.text.replace("\n",""))
    
all_data 
```

```python
all_data
```

```python
position=all_data[0::7]
position
```

```python
companies=all_data[1::7]
companies
```

```python
Market_Cap=all_data[2::7]

Market_Cap
  
```

```python
Market_Cap

Market_Cap1=[]
for i in Market_Cap :
  Market_Cap1.append(i.strip())
Market_Cap1
  
```

```python
Stock=all_data[3::7]
Stock1=[]
for i in Stock :
  Stock1.append(i.strip())
Stock1
```

```python
Country=all_data[4::7]
Country1=[]
for i in Country :
  Country1.append(i.strip())
Country1
```

```python
Sector=all_data[5::7]
Sector1=[]
for i in Sector :
  Sector1.append(i.strip())
Sector1
```

```python
Industry=all_data[6::7]
Industry1=[]
for i in Industry :
  Industry1.append(i.strip())
Industry1  
```

```python
import pandas as pd
top_companies=pd.DataFrame({})
top_companies["RANK"]=position
top_companies["COMPANY"]=companies
top_companies["MARCKET CAP"]=Market_Cap1
top_companies["STOCK"]=Stock1
top_companies["COUNTRY"]=Country1
top_companies["SECTOR"]=Sector1
top_companies["INDUSTREY"]=Industry1
```

```python
top_companies
```

```python
top_companies.to_csv('rutuja')
```
